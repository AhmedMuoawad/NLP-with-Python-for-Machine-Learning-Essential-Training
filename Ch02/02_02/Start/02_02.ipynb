### Read in raw text

```python
import pandas as pd
import re
import string
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']

data.head()
```

### Clean up text

```python
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text

data['body_text_nostop'] = data['body_text'].apply(lambda x: clean_text(x.lower()))
data.head()
```

### Stem text

```python
def stemming(tokenize_list):
    text = [ps.stem(word) for word in tokenize_list]
    return text

data['body_text_stemmed'] = data['body_text_nostop'].apply(lambda x: stemming(x))

data.head()
```

```python
import nltk 

wn = nltk.WordNetLemmatizer()

def lemmatizing(tokenize_list):
    text = [wn.lemmatize(word) for word in tokenize_list]
    return text

data['body_text_limmatize'] = data['body_text_nostop'].apply(lambda x: lemmatizing(x))

data.head()
```
