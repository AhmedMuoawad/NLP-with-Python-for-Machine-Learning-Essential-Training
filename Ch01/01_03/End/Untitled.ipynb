```python
# Read in the raw text 
rawData = open("SMSSpamCollection.tsv").read()

# Print the raw data 
rawData[0:500]
```

```python
# Handling the raw Data
parsedData = rawData.replace('\t', '\n').split('\n')
parsedData[0:500]
```

```python
# Create Data Frame
label = parsedData[0::2]
text = parsedData[1::2]

import pandas as pd 

df = pd.DataFrame({
                    'label' : label[0:-1],
                    'text' : text
    
                    })
df.head()
```
